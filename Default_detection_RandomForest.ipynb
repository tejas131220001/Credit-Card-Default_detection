```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('UCI_Credit_Card.csv')
```

```python
df.head()
```

```python
col=list(df.columns)
col[-1]='default'
df.columns=col
```

```python
df.head()
```

```python
  df.describe().T
```

```python
df.info()
```

```python
df['default'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('default',axis=1)
Y=df['default']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
from sklearn import metrics
y_pred=rfc.predict(X_test)
rfc.score(X_test,Y_test)
```

```python
print(metrics.classification_report(Y_test,y_pred))
```

```python
cm=metrics.confusion_matrix(Y_test,y_pred)
labels=['not defaluted','defaulted']
disp=metrics.ConfusionMatrixDisplay(cm,display_labels=labels)
disp.plot()
```

#output is highly imbalanced towards 0
#thatswhy we have very less precision,recall for output 1 category
# **We will upsample the imbalanced data**

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=67)
X_res,Y_res=sm.fit_resample(X,Y)
```

```python
Y_res.value_counts()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_res,Y_res,test_size=0.3)
rfc=RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
y_pred=rfc.predict(X_test)
rfc.score(X_test,Y_test)
```

```python
print(metrics.classification_report(Y_test,y_pred))
```

```python
cm=metrics.confusion_matrix(Y_test,y_pred)
labels=['not defaluted','defaulted']
disp=metrics.ConfusionMatrixDisplay(cm,display_labels=labels)
disp.plot()
```

```python
from sklearn.model_selection import GridSearchCV
params={'max_depth':range(2,30,5)}
rfc=RandomForestClassifier()
model_cv=GridSearchCV(rfc,param_grid=params,return_train_score=True,verbose=1,scoring='accuracy',cv=5,n_jobs=-1)
model_cv.fit(X_train,Y_train)
```

```python
results=pd.DataFrame(model_cv.cv_results_)
results
```

```python
sns.lineplot(x=results['param_max_depth'],y=results['mean_train_score'])
sns.lineplot(x=results['param_max_depth'],y=results['mean_test_score'])
plt.grid()
```

```python
#max depth =22
params={'n_estimators':range(200,2001,300)}
rfc=RandomForestClassifier(max_depth=15)
model_cv=GridSearchCV(rfc,param_grid=params,return_train_score=True,cv=5,scoring='accuracy',n_jobs=-1)
model_cv.fit(X_train,Y_train)
```

```python
results=pd.DataFrame(model_cv.cv_results_)
results
```
